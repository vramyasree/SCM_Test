```python
#Missing values
df = (df_1 .replace(r'^\s*$', np.nan, regex=True))
df_missing = df.isnull().sum().to_frame().rename(columns = {0 : "sumofmissingvalues"}).sort_values(by = "sumofmissingvalues", ascending = False)

df_missing ["total_rows"] = len(df)
df_missing_p = df.isnull().sum().to_frame().rename(columns = {0 : "%missingvalues"}).sort_values(by = "%missingvalues", ascending = False)*100/len(df)
df_m = pd.merge(df_missing, df_missing_p, left_index=True, right_index=True)
```

```python
#company code comparisions
df["Company_code_consistency"]= np.where(df['Company_Code_RPT']==df['Company_Code_IMT_C1'], 
                                           'match', 'mismatch')
df["Create_on_consistency"]= np.where(df['PO_Created_Date_RPT']==df['Created_On_IMT_C1'], 
                                           'match', 'mismatch')
df["Cost_Center_consistency"]= np.where(df['CostCenter_Code_RPT']==df['Cost_Center_IMT_C1'], 
                                           'match', 'mismatch')
df["Requestor_consistency"]= np.where(df['Request_User_ID_RPT']==df['Requisitioner_IMT_C1'], 
                                           'match', 'mismatch')
df["WBS_consistency"]= np.where(df['WBS_Code_RPT']==df['WBS_Element_IMT_C1'], 
                                           'match', 'mismatch')
df["GL_Account_consistency"]= np.where(df['GL_Account_RPT']==df['GL_Account_IMT_C1'], 
                                           'match', 'mismatch')
df["Profit_Center_consistency"]= np.where(df['Profit_Center_RPT']==df['Profit_Center_IMT_C1'], 
                                           'match', 'mismatch')
```

```python
df["Company_code_completeness"]= np.where((((df['Company_Code_RPT'].isnull()==True)&( df['Company_Code_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["Create_on_completeness"]= np.where((((df['PO_Created_Date_RPT'].isnull()==True)&( df['Created_On_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["Cost_Center_completeness"]= np.where((((df['CostCenter_Code_RPT'].isnull()==True)&( df['Cost_Center_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["Requestor_completeness"]= np.where((((df['Request_User_ID_RPT'].isnull()==True)&( df['Requisitioner_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["WBS_completeness"]= np.where((((df['WBS_Code_RPT'].isnull()==True)&( df['WBS_Element_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["GL_Account_completeness"]= np.where((((df['GL_Account_RPT'].isnull()==True)&( df['GL_Account_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')

df["Profit_Center_completeness"]= np.where((((df['Profit_Center_RPT'].isnull()==True)&( df['Profit_Center_IMT_C1'].isnull()!=True))),'missing_in_rpt', 'not_missing_in_rpt')
```

```python
df["Company_code_accuracy"]= np.where(((df['Company_code_completeness']=="not_missing_in_rpt")&(df['Company_Code_RPT']!=df['Company_Code_IMT_C1'])),'not_accurate', 'missing/accurate')

df["Create_on_accuracy"]= np.where(((df['Create_on_completeness']=="not_missing_in_rpt")&(df['PO_Created_Date_RPT']!=df['Created_On_IMT_C1'])),'not_accurate', 'missing/accurate')

df["Cost_Center_accuracy"]= np.where(((df['Cost_Center_completeness']=="not_missing_in_rpt")&(df['CostCenter_Code_RPT']!=df['Cost_Center_IMT_C1'])),'not_accurate', 'missing/accurate')

df["Requestor_accuracy"]= np.where(((df['Request_User_ID_RPT']=="not_missing_in_rpt")&(df['Request_User_ID_RPT']!=df['Requisitioner_IMT_C1'])),'not_accurate', 'missing/accurate')

df["WBS_accuracy"]= np.where(((df['Management_WBS_Code_RPT']=="not_missing_in_rpt")&(df['WBS_Code_RPT']!=df['WBS_Element_IMT_C1'])),'not_accurate', 'missing/accurate')

df["GL_Account_accuracy"]= np.where(((df['GL_Account_RPT']=="not_missing_in_rpt")&(df['GL_Account_RPT']!=df['GL_Account_IMT_C1'])),'not_accurate', 'missing/accurate')

df["Profit_Center_accuracy"]= np.where(((df['Profit_Center_RPT']=="not_missing_in_rpt")&(df['Profit_Center_RPT']!=df['Profit_Center_IMT_C1'])),'not_accurate', 'missing/accurate')
```

```python
Consistency_list=[]

for items in df.columns:
    if "consistency" in items:
        Consistency_list.append(items)
       
df_AUX =df[Consistency_list]
df_AUX = df_AUX.replace("mismatch",np.nan)


df_aux1 = df_AUX.isna().sum().to_frame().rename(columns = {0 : "mismatch_sum"}).sort_values(by = "mismatch_sum", ascending = False)

df_aux1 ["total_rows"] = len(df_AUX)
df_aux2 = df_AUX.isnull().sum().to_frame().rename(columns = {0 : "%mismatch"}).sort_values(by = "%mismatch", ascending = False)*100/len(df_AUX)
df_Consistency = pd.merge(df_aux1, df_aux2, left_index=True, right_index=True)

#df_m["Table_Name"] = df_EDL_1.columns.to_list()[1].split(".")[0]
df_Consistency.to_excel("df_Consistency_RPT_C1.xlsx")
df_Consistency
```

```python
Completeness_list=[]

for items in df.columns:
    if "completeness" in items:
        Completeness_list.append(items)
       
df_AUX =df[Completeness_list]
df_AUX = df_AUX.replace("missing_in_rpt",np.nan)


df_aux1 = df_AUX.isna().sum().to_frame().rename(columns = {0 : "missing_in_rpt_sum"}).sort_values(by = "missing_in_rpt_sum", ascending = False)

df_aux1 ["total_rows"] = len(df_AUX)
df_aux2 = df_AUX.isnull().sum().to_frame().rename(columns = {0 : "%missing_in_rpt"}).sort_values(by = "%missing_in_rpt", ascending = False)*100/len(df_AUX)
df_Completeness= pd.merge(df_aux1, df_aux2, left_index=True, right_index=True)

#df_m["Table_Name"] = df_EDL_1.columns.to_list()[1].split(".")[0]
df_Completeness.to_excel("df_Completeness_RPT_C1.xlsx")
df_Completeness
```

```python
Accuracy_list=[]

for items in df.columns:
    if "accuracy" in items:
        Accuracy_list.append(items)
       
df_AUX =df[Accuracy_list]
df_AUX = df_AUX.replace("not_accurate",np.nan)

df_aux1 = df_AUX.isna().sum().to_frame().rename(columns = {0 : "not_accurate_sum"}).sort_values(by = "not_accurate_sum", ascending = False)

df_aux1 ["total_rows"] = len(df_AUX)
df_aux2 = df_AUX.isnull().sum().to_frame().rename(columns = {0 : "%not_accurate"}).sort_values(by = "%not_accurate", ascending = False)*100/len(df_AUX)
df_Accuracy= pd.merge(df_aux1, df_aux2, left_index=True, right_index=True)

#df_m["Table_Name"] = df_EDL_1.columns.to_list()[1].split(".")[0]
df_Accuracy.to_excel("df_Accuracy_RPT_C1.xlsx")
df_Accuracy
```
